```
#chcemy pogrupowac po marce pojazdu, znalezc wartosc srednia (mean)
df.groupby('param_marka-pojazdu')['price_value'].agg(np.mean).plot(kind='bar')
```

```
#pracujemy, aby wykres byl bardziej czytelny
(
  df
  .groupby('param_marka-pojazdu')['price_value']
  .agg([np.mean, np.median, np.size])
  .sort_values(by='mean',ascending=False)
#sort_values(by='mean' -sortuj po wartosci sredniej
#(ascending=False) sprawia, ze mamy od najwyjszych na poczatku wykresu 
  .head(50)
#za duzo danych, ograniczamy do 50 
 
#teraz chcemy wydluzyc wykres 
).plot(kind='bar', figsize=(15,5))
```
